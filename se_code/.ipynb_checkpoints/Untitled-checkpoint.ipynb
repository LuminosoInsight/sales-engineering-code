```python
client = LuminosoClient.connect('/')
```

```python
client = client.change_path('projects/d53m338v/prs7rndc')
```

```python
subsets = client.get('subsets/stats')
```

```python
subsets
```

```python
subsets_to_remove = []
for subset in subsets:
    if subset['count'] < 150:
        subsets_to_remove.append(subset['subset'])

docs = []
while True:
    new_docs = client.get('docs', limit=25000, offset=len(docs))
    if new_docs:
        docs.extend(new_docs)
    else:
        break
```

```python
subsets_to_remove[16]
```

```python
for doc in docs:
    if subsets_to_remove[16] in doc['subsets']:
        print(doc)
        break
```

```python
for doc in docs:
    subset = []
    for doc_subset in doc['subsets']:
        if doc_subset not in subsets_to_remove:
            subset.append(doc_subset)
    doc['subsets'] = subset
        
```

```python
for doc in docs:
    if doc['_id'] == 'uuid-0b368961ec0c42e29a2ba66dab25b88b':
        print(doc)
        break
```

```python
batch = 0
batch_size = 10000
total_size = len(docs)
while batch < total_size:
    print(batch)
    end = min(total_size, batch + batch_size)
    docs_to_modify = docs[batch:end]
    client.put('docs', docs=docs_to_modify)
    print("Put in the docs to be modified")
    job_id = client.post('docs/recalculate')
    print("Recalculating")
    client.wait_for(job_id)
    batch += batch_size
```
